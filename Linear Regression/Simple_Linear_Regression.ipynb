```python
# Creating a dataset

import numpy as np

# Setting the random seed for reproducibility
np.random.seed(42)

# Generating the independent variable (X)
X = np.linspace(0,10,100)  # Generates 100 evenly spaced points between 0 and 10

# Generate the dependent variable (y) with some noise
noise = np.random.normal(0,10,100)  # Generates 100 random values from a normal distribution

# Linear relationship
y = 2*X + 3 + noise 
```

```python
# Creating a dataframe
import pandas as pd

data = pd.DataFrame({'X':X, 'y':y})
data.head()
```

```python
# Checking the data visually
import matplotlib.pyplot as plt

plt.scatter(X,y)
plt.xlabel("X")
plt.ylabel("y")
plt.title("Simple Linear Regression Data")
plt.show()
```

```python
# Regression Plot
import seaborn as sns

sns.regplot(data=data, x=data.X,y=data.y)
```

```python
# Fitting the simple linear regression to the data
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(data[['X']],data[['y']])
```

```python
# Print the coefficient and intercept

print("Coefficient:", lr.coef_[0])
print("Intercept:", lr.intercept_)
```
